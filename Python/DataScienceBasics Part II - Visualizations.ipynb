## Time series data with different variables

```python

import pandas as pd

emissions = pd.read_csv('C:/Users/Aa/Desktop/Coding_Practice/Python/datasets/Global_CO2_levels.csv')

grouped_emissions = emissions.groupby('Year').mean().sort_values('Year')
display(grouped_emissions.head())

grouped_emissions['pct_change'] = grouped_emissions['Seasonally Adjusted CO2 Fit (ppm)'].pct_change() *100
# print(grouped_emissions['pct_change'])

fig, ax = plt.subplots()

ax.plot(grouped_emissions.index, grouped_emissions['Seasonally Adjusted CO2 Fit (ppm)'], color='blue')
ax.set_xlabel('Year')
ax.set_ylabel('Seasonally Adjusted Mean CO2 concentration per year', color='blue')
ax.tick_params('y',colors='blue')

ax2 = ax.twinx()

ax2.plot(grouped_emissions.index, grouped_emissions['pct_change'], color='red')
ax2.set_ylabel('% change per year', color='red')
ax2.tick_params('y',colors='red')

ax.set_title('Mean CO2 growth')
plt.show() # would work better if relative temp would've been taken
```

## Function to plot time-series

### Annotating data

```python
def plot_timeseries(axes,x,y,color,xlabel,ylabel):
    axes.plot(x,y,color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y',colors=color)

fig, ax = plt.subplots()

plot_timeseries(ax,grouped_emissions.index, grouped_emissions['Seasonally Adjusted CO2 Fit (ppm)'],'blue','Year','Seasonally Adjusted Mean CO2 concentration per year')

ax2 = ax.twinx()

plot_timeseries(ax2,grouped_emissions.index, grouped_emissions['pct_change'],'red','Year','% change per year')

#Annotating time-series data
# annotating the point where pct change is > 0.8

#xy is the coordinates, xytext is used to position the overlapping text, arrowprops used for customizing the arrow
# ax2.annotate(">0.8 %", xy=[2016, 0.81], 
#              xytext=(2010,0.2),
#             arrowprops={"arrowstyle":"->", "color":"gray"}) 

ax2.annotate(">0.8 %", xy=[2016, 0.81], 
             xytext=(1960,0.8),
            arrowprops={"arrowstyle":"->", "color":"green"}) 

plt.show()
```

# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Styling plots

```python

sns.set_style('whitegrid')
sns.set_style('ticks')
sns.set_style('dark')
sns.set_style('darkgrid')

#Diverging Palettes
sns.set_palette("RdBu") # PRGn, RdBu_r

#Sequential Palettes
sns.set_palette('Blues') #Greys,PuRd,GnBu

#Context
sns.set_context('talk') #Poster,text,paper
```

## Scatterplot

```python
height = [62,64,69,75,66,68,65,71,76,73]
weight = [120,136,148,175,137,165,154,172,200,187]

#Scatterplot
g = sns.scatterplot(x=height, y=weight)

print(type(g)) # g is a axesSubplot object

g.set_title("Scatter Plot",y=1.03)
g.set(xlabel="Height",ylabel="Weight")
plt.show()
```

## Countplot

```python
gender = ["Female","Female","Female","Female","Male","Male","Male","Male","Male","Male"]

g = sns.countplot(x=gender)
g.set_title("Count Plot",y=1.03)
g.set(xlabel="Height",ylabel="Weight")

plt.show()
```

## Using hue for colors

```python
g1 = sns.scatterplot(x=height,y=weight,hue=gender)
g1.set_title("Scatter Plot with hue",y=1.03)
g.set(xlabel="Height",ylabel="Weight")

plt.show()

g2 = sns.scatterplot(x=height,y=weight,hue=gender, hue_order=["Male","Female"]) #Using hue_order to change order of color scheme
g2.set_title("Scatter Plot with hue and hue_order",y=1.03)
g.set(xlabel="Height",ylabel="Weight")

plt.show()
```

## Using palette

```python
palette_colors = {"Female": "red","Male":"black"}
g1 = sns.scatterplot(x=height,y=weight,hue=gender, hue_order=["Male","Female"], palette=palette_colors) #Using hue_order to change order of color scheme

g1.set_title("Scatter Plot with palette",y=1.03)
g.set(xlabel="Height",ylabel="Weight")

plt.show()
```

## relplot() - relational plots - subplots in a single figure

```python
df1 = pd.concat([pd.DataFrame(height, columns=['height'])], ignore_index=True)
df1['weight'] = weight
df1['gender'] = gender

g = sns.relplot(x="height",
            y="weight",
            data=df1,
            col="gender", #either use row or col for a [1,2] or a [2,1] figure, or use both for a [2,2] figure
#             row="gender",
#             col_wrap=2, to specify how many cols per row
            col_order=["Male","Female"], #Works like hue_order
#             style="size", # this is used to differentiate the point styles wrt various parameters
#             size = "size",
#             hue="size",
            alpha=0.8, # used for overlapping points
            kind="scatter")


print(type(g)) # g is a FacetGrid object


# g.fig.suptitle("TITLE",y=1.05)

g.set_titles("For {col_name}") #For both at once!
g.set(xlabel="Height",ylabel="Weight")

plt.show()
```
