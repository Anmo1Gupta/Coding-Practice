```python
story = urlopen('http://sixty-north.com/c/t.txt')
story_words = []
for line in story:
#     line_words = line.split()
    line_words = line.decode('utf-8').split()
    for word in line_words:
        story_words.append(word)
```

```python
# Nth root
def nth_root(radicand, n):
    return radicand**(1/n)
print(nth_root(16,2))
print(nth_root(27,3))
```

```python
def ordinal_suffix(value):
    s = str(value)
    if s.endswith('11'):
        return 'th'
    elif s.endswith('12'):
        return 'th'
    elif s.endswith('13'):
        return 'th'
    elif s.endswith('1'):
        return 'st'
    elif s.endswith('2'):
        return 'nd'
    elif s.endswith('3'):
        return 'rd'
    # else:
    #     return 'th'
    return 'th'

def ordinal_value(value):
    return str(value) + ordinal_suffix(value)

def display_nth_root(radicand, n):
    root = nth_root(radicand, n)
    message = "The " + ordinal_value(n) + " root of " + str(radicand) + " is: " + str(root)
    print(message)
```

```python
display_nth_root(64,4)
```

```python
#Random program to climb the Empire State Building
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
print(np.random.randint(1,7)) #To simulate a die
print(np.random.randint(1,7))

# Starting step
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1) # max() is used so that the step doesn't go into -ve
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 : # 0.1% chance that we fall all the way to the bottom
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]
print(ends)
# Plot histogram of ends, display plot
plt.clf()
plt.hist(ends, bins=10)
plt.show()
```
